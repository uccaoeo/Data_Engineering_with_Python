### Importing a SINGLE VALUE INTO THE DATABASE

```python
query = "insert into users (id,name,street,city,zip)\
values({},'{}','{}','{}','{}')"\ #using this method, you have to use quote to
#distinguish if datatype is string or numeric with the quote ({}==> numeric,'{}'==> string)
.format(1,'Big Bird','Sesame Street','Fakeville','12345')
```

```python
cur.mogrify(query)
```

### Importing numerous records into the database

```python
query2 = "insert into users (id,name,street,city,zip)\
values(%s,%s,%s,%s,%s)"   # psycopg2 guess the data type in this second method

data=(1,'Big Bird','Sesame Street','Fakeville','12345')
```

```python
cur.mogrify(query2, data) # Use mogrify to test what the query will send to the database
```

```python
cur.execute(query2, data) #execute the query
```

```python
conn.commit() #this code sends the executed query to the database
```

#### Using the faker library to impute multiple records to the PgAdmin Database

```python
fake = Faker() #initialize the Faker module
data = []
i=2
```

```python
for r in range(1000):
    data.append((i,fake.name(), fake.street_address(), fake.city(), fake.zipcode())) #create a tuple
    i+=1
```

```python
data_for_db = tuple(data) #convert array to tuple..making it thereby tuple of tuples
```

```python
query3 = "insert into users (id,name,street,city,zip)\
values(%s,%s,%s,%s,%s)"
```

```python
print(cur.mogrify(query3, data_for_db[400]))
```

```python
cur.executemany(query3, data_for_db)
conn.commit()
```
