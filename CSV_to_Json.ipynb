```python
def CsvToJson():
    for i,r in df2.iterrows():
        print(r['name'])
    df2.to_json('fromairflow.json', orient='records')
```

```python
default_args = {
    'owner': 'horlad',
    'start_date':dt.datetime(2022,5,22),
    'retries':1,
    'retry_delay':dt.datetime(minutes=2),
}
```

```python
with DAG('myairflowdag',
         default_args = default_args,
         schedule_interval =dt.datetime(minutes=2),
         
) as dag:
print_starting = BashOperator(task_id = 'starting',
                                 bash_command='echo "I am reading the CSV now..."')
CSVJson = PythonOperator(task_id='convertCSVtoJson',python_callable=CsvToJson)

print_starting >> CSVJson
```
